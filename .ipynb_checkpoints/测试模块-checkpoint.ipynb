### cv2颜色转换

```python
import hashlib
import lmdb
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data_dir="F:/Python_proj/data/vid15rpn_finetune"
db = lmdb.open(data_dir + '.lmdb', readonly=True, map_size=int(5e8))
```

```python
def imread(path):
    key = hashlib.md5(path.encode()).digest()
    img_buffer = txn.get(key)
    img_buffer = np.frombuffer(img_buffer, np.uint8)
    img = cv2.imdecode(img_buffer, cv2.IMREAD_COLOR)
    return img
```

```python
txn = db.begin(write=False)
```

```python
exemplar_name="F:/Python_proj/data/vid15rpn_finetune\\ILSVRC2015_train_00001000\\000044.00.x_75.52_52.02.jpg"
exemplar_img = imread(exemplar_name)
exemplar_img.dtype
```

```python
plt.imshow(exemplar_img)
```

```python
exemplar_img = cv2.cvtColor(exemplar_img, cv2.COLOR_BGR2RGB)# ndarray
```

```python
plt.imshow(exemplar_img)
```

### 创建汉宁窗

```python
size=(17,17,3)
cos_window = np.hanning(int(size[0]))[:, np.newaxis].dot(np.hanning(int(size[1]))[np.newaxis, :])
```

```python
cos_window = cos_window.astype(np.float32)
cos_window /= np.sum(cos_window)
```

```python
cos_window.shape
```

```python
plt.imshow(cos_window)
```

```python
b=cv2.resize(cos_window,(255,255))
```

```python
plt.imshow(b)
```
